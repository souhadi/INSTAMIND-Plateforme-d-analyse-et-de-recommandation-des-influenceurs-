```python
import pandas as pd

# Exemple de donnÃ©es
d=data['content_post']

# Convertir les donnÃ©es en DataFrame
df = pd.DataFrame(d)

# SÃ©parer les posts dans une colonne distincte
df['posts'] = df['content_post'].str.split("ago\n")

# CrÃ©er des colonnes distinctes pour chaque post
for i in range(len(df['posts'].iloc[0])):
    df[f'post_{i+1}'] = df['posts'].apply(lambda x: x[i] if len(x) > i else '')

# Supprimer les colonnes inutiles
df.drop(['content_post', 'posts'], axis=1, inplace=True)

# Afficher le DataFrame rÃ©sultant
df
```
