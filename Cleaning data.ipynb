```python
from langdetect import detect
import pandas as pd
import spacy
import re
```

```python
# Étape 1 : Chargement du fichier CSV
file_path = '/influenceurs.csv'
df = pd.read_csv(file_path, sep='\t')
```

```python

# Liste des colonnes à convertir en minuscules
colonnes_texte = ['user_name', 'profile_desc','post_1','post_2','post_3','post_4','post_5','post_6','post_7','post_8','post_9','post_10','post_11','post_12']

# Convertir les colonnes en minuscules
for colonne in colonnes_texte:
    df[colonne] = df[colonne].str.lower()

# Afficher le DataFrame avec les colonnes en minuscules
df
```

```python

# Définir une fonction pour supprimer la mise en forme en gras
def supprimer_mise_en_forme_gras(texte):
    return re.sub(r'\*\*(.*?)\*\*', r'\1', texte)

for colonne in colonnes_texte:
    df[colonne] = df[colonne].apply(supprimer_mise_en_forme_gras)

df
```

```python

# Colonnes de texte à traiter
text_columns = ['profile_desc', 'post_1', 'post_2', 'post_3', 'post_4', 'post_5', 'post_6', 'post_7', 'post_8', 'post_9', 'post_10', 'post_11', 'post_12']

# Définir la fonction de nettoyage
def clean_text(text):
    # Supprimer les caractères spéciaux, les chiffres et les espaces supplémentaires
    text = re.sub(r"[^\w\s]|[\d]", "", text)
    # Supprimer les espaces supplémentaires
    text = ' '.join(text.split())
    return text

# Appliquer le nettoyage à chaque colonne de texte
for colonne in text_columns:
    df[colonne] = df[colonne].apply(clean_text)

```

```python
# Étape 2 : Nettoyage linguistique
nlp = spacy.load('xx_ent_wiki_sm')

def clean_single_text(text):
    doc = nlp(text)
    clean_text = ' '.join([token.lemma_ for token in doc if not token.is_punct and not token.is_stop])
    return ' '.join([word for word in clean_text.split() if word.isalpha()])

```
